```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python

def boltzmann_distribution(beta, heat):
    return np.exp(-(beta * heat))
```

```python

betas = np.linspace(0, 1, 5)

x_vals = np.linspace(0, 10, 100)

for beta in betas:
    plt.plot(x_vals, boltzmann_distribution(beta, x_vals), label=f"""beta: {beta}""")


plt.legend()
plt.show()
```
